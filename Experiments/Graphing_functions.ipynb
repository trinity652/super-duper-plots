```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

def plot_function():
    x = sp.symbols('x')
    expr_input = input("Enter a function in terms of x (e.g., x**2 + 3*x - 4): ")
    
    try:
        # Convert the string input to a symbolic expression
        expr = sp.sympify(expr_input)
        
        # Lambda function to convert symbolic to numerical computation
        f = sp.lambdify(x, expr, "numpy")
        
        # Generate x and y values for the plot
        x_vals = np.linspace(-10, 10, 400)
        y_vals = f(x_vals)

        # Plotting
        plt.plot(x_vals, y_vals, label=str(expr))
        plt.title(f"Graph of y = {expr}")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
        plt.legend()
        plt.show()
    
    except:
        print("There was an error parsing or plotting the function.")
```

```python
plot_function()
```
